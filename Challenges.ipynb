Recording Annual Rainfall with a Tuple


```python
#Data for india
GivenData = (750,760,770,740,720,730,750,760,740,725)

indexof2015 = 2

#correcting the value for 2015 using the average of adjacent years to correct it
value = (GivenData[1] + GivenData[3]) // 2
print(value)

correctdata = tuple(GivenData[:indexof2015] + (GivenData[indexof2015] + value,) + GivenData[indexof2015 + 1:])
print(correctdata)

 #new average
new_average = sum(correctdata) / len(correctdata)
print(new_average)
```

Tracking Emission levels with a dictionary  

```python
#emissions data in millions of metric tons
emissions_data = {
    'USA' : [5000,4950,4900,4850,4800,4750,4700,4650,4600,4550],
    'CHINA': [10000,10100,10200,10300,10400,10500,10600,10700,10800,10900],
    'INDIA':[3000,3050,3100,3150,3200,3250,3300,3350,3400,3450]
}

indexof2017 = 4

#correcting China data for 2017 from thousand to million
emissions_data['CHINA'][indexof2017] /= 0.001
# print(emissions_data['CHINA'][indexof2017])

#correcting 10600
indexposition = 6

value = (emissions_data['CHINA'][5] + emissions_data['CHINA'][7]) / 2
print(value)

# Correct the emissions data for China in 2017
emissions_data['CHINA'][indexposition] += value
print(emissions_data['CHINA'][indexposition])


 # new averages for each country
averages = {country: sum(data) / len(data) for country, data in emissions_data.items()}

# corrected data and new averages
print("Corrected Emissions Data:")
for country, data in emissions_data.items():
    print(f"{country}: {data}")

print("\nNew Averages:")
for country, average in averages.items():
    print(f"{country}: {average}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Corrected Glacier Volumes Data
years_glacier = list(range(2013, 2023))

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(GivenData, HimalayanPeak, c='blue', label='Rainfall vs. Glacier Volumes', marker='o')

# Fit a linear regression line
slope, intercept, r_value, p_value, std_err = linregress(GivenData, HimalayanPeak)
line = slope * np.array(GivenData) + intercept
plt.plot(GivenData, line, color='red', label='Trendline')

# Set labels and title
plt.xlabel('Rainfall (mm)')
plt.ylabel('Glacier Volumes (millions of metric tons)')
plt.title('Scatter Plot with Trendline: Rainfall vs. Glacier Volumes')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()

```
